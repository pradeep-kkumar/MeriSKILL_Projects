# Data Visualization

```python
plt.figure(figsize=(20,20))
sns.heatmap(hr_data.corr(),cmap='rainbow',annot=True,fmt='.3g',linecolor="k",linewidths=2)
plt.show()
```

```python
sns.countplot(x='Department',data=hr_data)
plt.title('Count of employees of each Department')
plt.xticks(rotation=45)
plt.xlabel('Department')
plt.ylabel('Count')
plt.show
```

```python
sns.countplot(x='JobRole',data=hr_data)
plt.title('Job Role')
plt.xticks(rotation=45)
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.show
```

```python
sns.barplot(x=hr_data['OverTime'],y=hr_data['Age'])
plt.show()
```

```python
sns.histplot(hr_data["OverTime"])
plt.show()
```

```python
sns.histplot(hr_data["MaritalStatus"])
plt.show()
```

```python
data = {'Gender': hr_data['Gender']}
df = pd.DataFrame(data)

gender_counts = df['Gender'].value_counts()

# Create a pie chart
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)

plt.axis('equal') 
plt.title('Gender Distribution')
plt.show()
```

```python
hr_data['EducationField'].value_counts().plot(kind="bar")

# Optionally, you can add labels and a title
plt.xlabel("Education Field")
plt.ylabel("Count")
plt.title("Count of each Education Field")
```

```python
hr_data['Department'].value_counts().plot(kind="bar")

# Optionally, you can add labels and a title
plt.xlabel("Department")
plt.ylabel("Count")
plt.title("Count of each Department")
```
